## *Importing the Libraries*

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## *Loading the Dataset*

```python
# Load the Iris dataset available through Seaborn library
df = sns.load_dataset('iris')
```

# *Exploratory Data Analysis (EDA)*

## *Scatterplots*

```python
# Plot scatter plots to analyze the relationships between the variables 
plt.figsize = (10,8)
sns.set_style('whitegrid')
sns.pairplot(df, hue = 'species')
plt.suptitle('Scatter Plots of Iris Features', y=1.02, weight='bold', style ='italic')
plt.show()
```

## *Histograms*

```python
# Drop 'species' to get only numeric features
features = df.drop(columns=['species'])

# Create subplots for the histograms
fig, axes = plt.subplots(2, 2, figsize=(10,8))
axes = axes.flatten()

# Plot histograms for each feature by species to examine data distribution
for i, col in enumerate(features):
    sns.histplot(data = df, x = col, hue = 'species', kde = True, bins =20, ax= axes[i])
    axes[i].set_title(f'Distribution of {col}')
    
# Adjust layout
plt.tight_layout()
plt.suptitle('Distribution of Iris Features', y=1.02, weight='bold', style='italic')
plt.show()
```

## *Box Plots*

```python
# Create subplots for the box plats
fig, axes = plt.subplots(2,2,figsize=(10,8))
axes = axes.flatten()

# Plot box plots for each feature by species to detect outliers and spread of values
for i, col in enumerate(features):
    sns.boxplot(data=df, y=col, hue='species',ax=axes[i])
    axes[i].set_title(f'Boxplot of {col}')

# Adjust layout 
plt.tight_layout()
plt.suptitle('Box plots of Iris Features', y=1.02, weight='bold', style='italic')
plt.show()
```
